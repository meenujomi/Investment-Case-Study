## Step 4 : Country Analysis

```python
# Arranging all the countries according to their investment rates and finding the top 9
top9 = master_frame_final.pivot_table(values= 'raised_amount_usd', index = 'country_code', aggfunc = 'sum' ).sort_values(['raised_amount_usd'], ascending = False).iloc[:9]
top9['country_code'] = top9.index
top9.index = np.arange(1, len(top9)+1)
top9
```

### 4.1 Solving to get value to put in table 3.1

## Step 5 : Sector Analysis 1

```python
# Loading mapping file into 'mapping' dataframes
mapping = pd.read_csv('F:\\data science\\casestudy\\mapping.csv', encoding= 'ISO-8859-1')
mapping
```

```python
# Cleaning of mapping dataframe
mapping['category_list'] = mapping['category_list'].str.replace('0','na').str.lower()
mapping = mapping.drop(['Blanks'],axis =1)
mapping
```

```python
# Creating a new dataframe with the main sectors and their category list from mapping dataframe
mapping = mapping.melt(id_vars = ['category_list'], var_name ='main sector' )
mapping = mapping[mapping['value'] !=0]
mapping
```

```python
# Extracting the primary sector of each category list from the category_list column 
master_category = master_frame_final['category_list'].str.split(pat='|',n=1, expand=True)
master_category = master_category.iloc[:,0]
master_category = master_category.to_frame(name= 'primary sector')
master_category = master_category.drop_duplicates(keep='first')
master_category['category_list'] = master_frame.loc[ : , ['category_list']]
master_category['primary sector'] = master_category['primary sector'].str.lower()
master_category
```

```python
# Using mapping dataframe to map each primary sector to one of the main sectors and putting in 'others' to the ones that has none
sectors = pd.merge(mapping,master_category, how='right',left_on='category_list',right_on = 'primary sector')
sectors = sectors.drop(['category_list_x','value'],axis =1)
sectors['category_list_y'] = sectors['category_list_y'].str.lower()
sectors.loc[sectors['main sector'].isnull(), 'main sector'] = ['Others']
sectors
```

```python
# Merging master_Frame and sectors
master_frame_sector = pd.merge(master_frame_final,sectors, how='left',left_on='category_list',right_on = 'category_list_y')
master_frame_sector = master_frame_sector.drop(['category_list_y'],axis =1)
master_frame_sector
```

## Step 6 : Sector Analysis 2

```python
# A dataframe where country is USA
D1 = master_frame_sector.loc[master_frame_sector['country_code']== 'USA']
D1
```

```python
# A dataframe where country is United Kingdom
D2 = master_frame_sector.loc[master_frame_sector['country_code']== 'GBR']
D2
```

```python
# A dataframe where country is India
D3 = master_frame_sector.loc[master_frame_sector['country_code']== 'IND']
D3
```

```python
# Adding count and total investment to the above table for USA
sector_usa = D1.groupby('main sector')['raised_amount_usd'].count()
sector_usa = sector_usa.to_frame(name= 'count of investments')
sector1_usa = pd.merge(D1,sector_usa, how='left',on='main sector')
sector2_usa = sector1_usa.groupby('main sector')['raised_amount_usd'].sum()
sector2_usa = sector2_usa.to_frame(name= 'sum of investments')
D1 = pd.merge(sector1_usa,sector2_usa, how='left',on='main sector')
D1
```

```python
# Adding count and total investment to the above table for United Kingdom
sector_gbr = D2.groupby('main sector')['raised_amount_usd'].count()
sector_gbr = sector_gbr.to_frame(name= 'count of investments')
sector1_gbr = pd.merge(D2,sector_gbr, how='left',on='main sector')
sector2_gbr = sector1_gbr.groupby('main sector')['raised_amount_usd'].sum()
sector2_gbr = sector2_gbr.to_frame(name= 'sum of investments')
D2 = pd.merge(sector1_gbr,sector2_gbr, how='left',on='main sector')
D2
```

```python
# Adding count and total investment to the above table for India
sector_ind = D3.groupby('main sector')['raised_amount_usd'].count()
sector_ind = sector_ind.to_frame(name= 'count of investments')
sector1_ind = pd.merge(D3,sector_ind, how='left',on='main sector')
sector2_ind = sector1_ind.groupby('main sector')['raised_amount_usd'].sum()
sector2_ind = sector2_ind.to_frame(name= 'sum of investments')
D3 = pd.merge(sector1_ind,sector2_gbr, how='left',on='main sector')
D3
```

### 6.1 Solving to get value to put in table 5.1

```python
# Total count and sum of investments in USA respectively 
D1_sum_count = D1.pivot_table(values= ['sum of investments','count of investments'], index = 'main sector', aggfunc = 'mean' ).sort_values(['sum of investments'], ascending = False)
print(D1_sum_count['count of investments'].sum())
print(round(D1_sum_count['sum of investments'].sum(),2))
```

```python
# To find the top 3 main sectors (sum and count) in GBR
top_sector_GBR = D2_sum_count.sort_values(['count of investments'], ascending = False).iloc[:3]
top_sector_GBR
```

```python
# To get the hightest invest company (taking count)
D2.pivot_table(values= 'count of investments', index = ['main sector','name'], aggfunc = 'count').sort_values(['count of investments'],ascending=False)
```

```python
# Total count and sum of investments in IND respectively 
D3_sum_count = D3.pivot_table(values= ['sum of investments','count of investments'], index = 'main sector', aggfunc = 'mean' ).sort_values(['sum of investments'], ascending = False)
print(D3_sum_count['count of investments'].sum())
print(round(D3_sum_count['sum of investments'].sum(),2))
```
